### DATA CLEANING

```python
# I will begin by dropping all duplicates
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
#check if duplicates have been dropped
df.duplicated().sum()
```

```python
#check for rows with missing values
df.isnull().sum()
```

```python
# MOST OF THE MISSING VALUES ARE STRINGS
# so I WILL BE DROPPING THEM BECAUSE I CANNOT FILL IN VALUES

drop_rows = df.dropna( how='any', subset=['imdb_id', 'cast','homepage','director','tagline','keywords','overview','genres','production_companies'])
```

```python
#check null values 
df.isnull().sum()
```

```python
# i will be renaming the title column as name and then using it as the table index
df = df.rename(columns={'original_title':'name'})
#check if it worked
df.head(1)
```

```python
# set name as index
df.reset_index().set_index('name',inplace=True)

```

```python
df.head(1)
```

```python
#delete id column
df = df.drop('id', axis=1)
```

```python
df.head(1)
```

```python
#rename column 'vote_average ' to 'rating'
df = df.rename(columns={'vote_average':'rating'})
#check
df.head(1)
```

```python
# i will be removing rows which have 0 for either the 'budget_adj' or the 'revenue_adj' column

df = df[df['budget_adj'] > 0]
df = df[df['revenue_adj'] > 0]
```

```python
## check if there are still rows with 0 in budget_adj and revenue_adj

df.query("(budget_adj == 0) | (revenue_adj == 0)").shape
```

## Exploratory Data Analysis



### Research Question 1 : Do the best rated movies have higher profits?

```python
# WE ALREADY HAVE ACOLUMN FOR RATINGS SO WE NEED TO FIND THE PROFIT
# OF EACH MOVIES TO KNOW THE ANSWER TO THIS QUESTION

#TO FIND PROFITS WE SUBTRACT THE BUDGET FROM REVENUE AND ASSIGN A COLUMN TO IT

df.insert(3, column='profit', value = df['revenue']-df['budget'])
```

```python
# let us see our new coulmn

df.head()
```

```python
# i want the profit column to come before the budget and revenue
# so i have to delete this column and change the 3 to 2 in the new column

df.drop('profit', axis = 1, inplace = True)
```

```python
# create profit again
df.insert(2, column='profit', value = df['revenue']-df['budget'])

```

```python
#check
df.head(1)
```

# Research Question 2 : Do movies with higher budgets get higher ratings?

```python
# in order to solve this we will divide the budget into two groups
# Higher and lower. We will then proceed to check the average ratings
# for both these groups in order to get our answer

df['budget'].median()
```

```python
# select movies with budget less than median

low_budget= df.query('budget < 15000000')

#select movies with budget more than median

high_budget= df.query('budget >= 15000000')

# get the mean rating for low and high budget
low_budget.rating.mean(), high_budget.rating.mean()
```

### so we deduce that movies with higher budget get slightly higher ratings
